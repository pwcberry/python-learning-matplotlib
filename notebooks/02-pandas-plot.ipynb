# pandas: Basic plotting

The `pandas` library provides some basic plotting functionality.

```python
%matplotlib inline

import pandas as pd
import numpy as np

data = pd.read_csv('../data/Melbourne_Annual.csv')
```

```python
# Plot the annual mean temperature according to year
data.plot(x = 'Year', y = 'Temperature')
```

```python
# Plot the annual total rainfall according to year
data.plot(x = 'Year', y = 'Rainfall')
```

```python
# Plot the rainfall for last 20 years as a bar graph
# Add a label for the Y-axis

ax = data[-20:].plot.bar(x = 'Year', y = 'Rainfall', legend=False)
ax.set_ylabel('Rainfall - mm');
```

```python
# Plot the temperatures for the last 40 years with the long-term mean
# Let's create a new `DataFrame` view
last_forty = data[-40:]

# Add another column to this data frame to represent the long-term mean
# We'll use numpy to create a list containing one value and of the same length as the new `DataFrame` and will be the source of a new `Series`
mean_temp = np.full(40, data['Temperature'].mean())
last_forty.insert(loc=0, column='Mean_Temp', value=mean_temp)

# Plot
last_forty[['Year', 'Temperature', 'Mean_Temp']].plot(x = 'Year', legend=False)
```
